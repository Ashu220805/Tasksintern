```python
import pandas as pd
# Load the dataset
df = pd.read_csv('Titanic-Dataset.csv')  # Adjust the path if needed
print(df.isnull().sum())
```

```python
numeric_cols = df.select_dtypes(include='number').columns

# Calculate mean, median, and mode
mean_values = df[numeric_cols].mean()
median_values = df[numeric_cols].median()
mode_values = df[numeric_cols].mode().iloc[0]  # mode() can return multiple rows; take the first

# Display the results
print("ðŸ“Š Mean:\n", mean_values)
print("\nðŸ“Š Median:\n", median_values)
print("\nðŸ“Š Mode:\n", mode_values)
```

```python
# Summary statistics for numeric columns
print(df.describe())

# Summary statistics for categorical columns
print(df.describe(include='object'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
numeric_features = ['Age', 'Fare', 'SibSp', 'Parch']
df[numeric_features].hist(figsize=(10, 8), bins=20)
plt.tight_layout()
plt.show()

# Boxplots
for col in numeric_features:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
corr = df[numeric_features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

# Survival rate by class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Age distribution by survival
sns.histplot(data=df, x='Age', hue='Survived', bins=20, kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```
